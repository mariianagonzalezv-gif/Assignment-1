I reviewed one of the top solutions under the LeetCode "Solutions" tab: https://leetcode.com/problems/two-sum/solutions/3619262/3-method-s-c-java-python-beginner-friendly/
This solution showed me that I can use the function len to then tell the loop how many times to run (makes sure every number is read). Additionally I learned that I need to create a number map so that each number's index is remembered



```python
# Your draft code here (does not need to be fully correct)
class Solution:
  def twoSum(self, nums: List[int], target: int) -> List[int]:
    # make a dictionary to keep track of numbers weâ€™ve seen so far
    numberMap = {}

    # loop through the array using the length of nums
    for i in range len(nums):

      # find the other number we need by subtracting current number from target
      number_two = target - nums

      # if the number we need is already in the map, we found a pair
      if number_two in numberMap:
        return [i, numberMap[nums[i]]]

      # otherwise add the current number and its index into the map
      numberMap[nums[i]] = i

    # if nothing is found, just return an empty list
    return []
```
