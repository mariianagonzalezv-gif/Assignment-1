I learned how to go about this problem using queues. You check the student at the front and compare the sandwich preference. If they match, move on. If they don't, student goes to the back of the line. If you loop throughout the whole queue without serving, remaining students dont eat.



```python
class Solution:
  def countStudents(self, students: List[int], sandwiches: List[int]) -> int:
    students_queue = deque(students)
    sandwiches_list = list(sandwiches)

    skip_count = 0   # how many students we have moved without anyone eating

    while students_queue and sandwiches_list:
      current_student = students_queue[0]
      current_sandwich = sandwiches_list[0]

        # if the student likes the sandwich, they take it
      if current_student == current_sandwich:
        students_queue.popleft()
        sandwiches_list.pop(0)

      else:
            # move student to the back of the queue
        students_queue.append(students_queue.popleft())
        skip_count += 1

        # if we went through the whole queue without progress, stop
      if skip_count == len(students_queue):
        return

    # students left in the queue can't eat
    return len(students_queue)
```
